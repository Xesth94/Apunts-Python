1. **Comparison operations** compares some value or operand. Then based on some condition, they produce a Boolean (True or False). We use `==` to determine if two values are equal. Inequalities `< >` and `<= >=` are used to see if the values are higher or lower than a value of the variable. `!=` if 2 values are not equal this is True. Those commands are useful in strings too.

```python
a = 6

# "==" to determine if two values are equal
print(a ==7)

# "<" to check if the variable is lower than the value
print(a < 8)

# "!=" not equal
print(a!=6)
```

2. **Branching** allows us to run different statements for a different input. There are different statements: `if`, `else` and `elif`.

```python
# "if" Statement. If the condition is True prints 'you can enter'.
age = 19

if (age > 18):
    print('you can enter')
```

```python
# "else" Statement. If the condition is False, prints 'move on'.
age = 17

if (age > 18):
    print('you can enter')
    
else:
    print('move on')
```

```python
# "elif" is short for else if. Allows to check additional conditions if the preceding condition if false.
age = 18

if (age > 18):
    print('you can enter')

elif (age == 18):
    print('go to see Pink Floyd')
    
else:
    print('move on')
```

Logic operations take Boolean values and produce different Boolean values. The `not()` operator is False when the input is True. The `or` operator is only False if all the Boolean values are False. Analogously, `and` operator only gives a True if all the values are True.

```python
album_year = 1990

if (album_year < 1980) or (album_year > 1989):
    print("The album was made in the 70's or 90's")

else:
    print("The album was made in the 1980's")

# The logic operator is True, so prints the first line
```

```python
album_year = 1983

if (album_year > 1979) and (album_year < 1990):
    print("The album was made in the 1980's")

else:
    print("The album was made in the 70's or 90's")

# The logic operator is True, so prints the first line
```

* #### **Loops**

We will cover in particular `for` and `while` loops. 

Before talking about loops, let's see the `range()` function. If the input is a positive integer, the output is a sequence that contains the same number of elements as the input but starts at zero. If the input is 3, the output is the sequence [0,1,2]. If the `range()` function has two inputs (10,15), when the first input is smaller than the second, the output is a sequence that starts at the first input and the secuence iterates up to but not including the second number: [10,11,12,13,14]. In Python 3 this function won't display the lists mentioned in the prompt, but has the same utility in loops.

1. **For** loops

`For` loops perform a task over and over. We will focus on lists, but many of the procedures can be used also on tuples.

```python
squares = ["red", "yellow", "green", "purple", "blue"]

# Convert every color of squares in white squares using "range()".
for i in range(0, 5):
    squares[i]= "white"
print(squares)

# We don't need to use "range()" for this iteration:
for x in ['A','B','C']:
    print(x+'A')
```

Another useful function in iterations is `enumerate()`. It can be used to obtain the index and the element in the list.

The argument of the function `enumerate()` is the list. The variable **"i" is the index** and the variable **"x" is the corresponding element** in the list:

```python
squares = ["red", "yellow", "green", "purple", "blue"]

for i,x in enumerate(squares):
    print(i, x)
```

2. **While** loops

`While` loops are similar to `for` loops but instead of executing a statement a set number of times a while loop will only run if a condition is met.

```python
squares = ["red", "red", "red", "purple", "blue"]

Newsquares = [] # Create a new list of new squares
i = 0 # We start the index at 0. The while loop will be executed repeteadly until the condition inside the bracket is false

while(squares[i] == 'red'):
    Newsquares.append(squares[i])
    i = i+1
print(Newsquares)
print(i)
```
